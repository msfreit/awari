#### Tarefa 2. Descubra qual a idade média por ocupação.

*Dica: use a função groupby do Pandas na coluna occupation*

````python
(df
 .groupby("occupation")
 .agg({"age": "mean"}))
````

#### Tarefa 3. Para cada ocupação descubra a idade mínima e máxima.

````python
(df
 .groupby("occupation")
 .agg({"age": ["min", "max"]}))
````

````python
(df
 .groupby("occupation")
 .agg({"age": "describe"}))
````

#### Tarefa 4. Para cada combinação de ocupação e sexo, calcule a idade média.

Exemplo:

```
occupation     gender
administrator  F         40.638889
               M         37.162791
artist         F         30.307692
               M         32.333333
```

````python
(df
 .groupby(["occupation", "gender"])
 .agg({"age": "mean"}))
````

#### Tarefa 5. Para cada ocupação calcule a porcentagem de homens e mulheres.

Exemplo:

```
occupation     gender
administrator  F          45.569620
               M          54.430380
artist         F          46.428571
               M          53.571429
doctor         M         100.000000
educator       F          27.368421
               M          72.631579
```

````python
result = df.groupby(["occupation", "gender"]).agg({"user_id" : "count"})
result.groupby(level=0).apply(lambda x: 100 * x / x.sum())
````

````python
(df.groupby(['occupation', 'gender'])['gender']
 .size().groupby(level=0)
 .apply(lambda x: 100 * x / x.sum())
 .round(2))
````

````python
occupation_gender = (df
                     .groupby(["occupation", "gender"])
                     .agg({"user_id": "count"})) 
````

````python
occupation_total_users = df.groupby("occupation").agg({"user_id":"count"})
````

````python
occupation_gender / occupation_total_users
````
