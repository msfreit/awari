```python
#!pip install tensorflow
from tensorflow.keras.applications.resnet import (
    ResNet50,
    decode_predictions,
    preprocess_input,
)
from skimage.transform import resize
from skimage.io import imread
import matplotlib.pyplot as plt
```

```python
model = ResNet50(weights="imagenet")
```

```python
test_image = imread("https://s2.glbimg.com/mYgwlPa7vtIiUk6kROUxJUi2yyo=/0x0:620x413/984x0/smart/filters:strip_icc()/i.s3.glbimg.com/v1/AUTH_cf9d035bf26b4646b105bd958f32089d/internal_photos/bs/2020/a/4/Ik8J1fQYirf6wYRvRJ8Q/2020-03-20-novo-tracker-1.jpg")
```

```python
plt.imshow(test_image)
```

```python
resized_image = resize(test_image, (224,224), preserve_range=True)
preprocessed_input = preprocess_input(resized_image).reshape(1,224,224,3)
```

```python
prediction = model.predict(preprocessed_input)
```

```python
decode_predictions(prediction)
```
