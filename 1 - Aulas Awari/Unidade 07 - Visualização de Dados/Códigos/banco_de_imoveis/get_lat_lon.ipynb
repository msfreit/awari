## Exemplo de como utilizar o api do googlemaps para pegar a latitude e longitude de um endereço. Repare que este api te dá alguns créditos mensais para uso gratuito. Maiores detalhes de pricing na documentação: 

- https://console.cloud.google.com/marketplace/product/google/geolocation.googleapis.com
- Eu rodei para os quase 16 mil registros do nosso dataset e nada foi cobrado, mas sempre confira o pricing antes de usar o API

```python
df = pd.read_csv("imoveis_primary.csv")
```

```python
# Seguindo o que fizemos no office hours em que usamos os APIs do google
api_key = "SUA_API_KEY" #Adicione sua api_key. Voce precisa criar uma no google cloud
gmaps = googlemaps.Client(key=api_key)
api_results = []
```

```python
# Exemplo: Enviamos um endereço para o API e essa é a resposta:
gmaps.geocode(df["address"].iloc[0])
```

```python
# Rodando para todas as linhas
for i, row in df.iterrows():
    if i%100==0:
        print(i)
    _id = row["id"]
    result = gmaps.geocode(row["address"])
    api_results.append({"id": _id, "results": result})
```

```python
to_df = []
for i,r in enumerate(api_results):
    _id = r["id"]
    lat = None
    lon = None
    if len(r["results"])>0:
        lat = r["results"][0]["geometry"]["location"]["lat"]
        lon = r["results"][0]["geometry"]["location"]["lng"]
    to_df.append((_id, lat, lon))
```

```python
df = df.merge(pd.DataFrame(to_df, columns=["id", "lat", "lon"]), on="id", how="left")
```
