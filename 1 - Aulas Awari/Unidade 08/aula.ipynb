```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set() #muda o layout dos graficos
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/awarischool/EDA-exercises/master/titanic.csv")
df.head()
```

## histograma

```python
df["Age"].plot(kind="hist")
```

```python
# plt
plt.hist(df["Age"])
```

```python
sns.histplot(data=df, x="Age", hue="Sex", stat="count", kde=True,  color=sns.color_palette("rocket") )
```

```python
df.loc[df["Age"].between(20,30)].groupby("Sex").count()
```

## Scatterplot

```python
df.plot(kind="scatter", x="Age" ,y="Fare")
```

```python
plt.scatter(x=df["Age"],y=df["Fare"])
```

```python
sns.scatterplot(data=df, x="Age", y="Fare", hue="Survived")
```

```python
df["Age"].plot(kind="box")
```
